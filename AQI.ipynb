Air Pollution is contamination that modifies the natural characteristics of the atmosphere often comes from household combustion devices, motor vehicles, industrial facilities and forest fires. Several pollutants are implicated in health concerns include particulate matter, carbon monoxide, ozone, nitrogen dioxide and sulfur dioxide.

This dataset provides geolocated information about the following pollutants:

Nitrogen Dioxide [NO2] : Nitrogen Dioxide is one of the pollutants coming from cars, trucks and buses emissions, power plants and off-road equipment. Exposure over short periods can aggravate respiratory diseases, like asthma. Longer exposures may contribute to develoment of asthma and respiratory infections. People with asthma, children and the elderly are at greater risk for the health effects of NO2.

Ozone [O3] : The Ozone molecule is harmful for outdoor air quality. It is created by chemical reactions between oxides of nitrogen and volatile organic compounds (VOC). Differently from the good ozone located in the upper atmosphere, ground level ozone can provoke several health problems like chest pain, coughing, throat irritation and airway inflammation. Furthermore it can reduce lung function and worsen bronchitis, emphysema, and asthma. Ozone affects also vegetation and ecosystems. In particular, it damages sensitive vegetation during the growing season.

Carbon Monoxide [CO] : Carbon Monoxide is a colorless and odorless gas. Outdoor, it is emitted in the air above all by cars, trucks and other vehicles or machineries that burn fossil fuels. Such items like kerosene and gas space heaters, gas stoves also release CO affecting indoor air quality. Breathing air with a high concentration of CO reduces the amount of oxygen that can be transported in the blood stream to critical organs like the heart and brain. At very high levels, which are not likely to occur outdoor but which are possible in enclosed environments. CO can cause dizziness, confusion, unconsciousness and death.

Particulate Matter [PM2.5] : Atmospheric Particulate Matter, also known as atmospheric aerosol particles, are complex mixtures of small solid and liquid matter that get into the air. If inhaled they can cause serious heart and lungs problem. They have been classified as group 1 carcinogen by the International Agengy for Research on Cancer (IARC). PM10 refers to those particules with a diameter of 10 micrometers or less. PM2.5 refers to those particles with a diameter of 2.5 micrometers or less.

Here are the descriptions of this dataset:
- `Country` : Name of the country
- `City` : Name of the city
- `AQI Value` : Overall AQI value of the city
- `AQI Category` : Overall AQI category of the city
- `CO AQI Value` : AQI value of Carbon Monoxide of the city
- `CO AQI Category` : AQI category of Carbon Monoxide of the city
- `Ozone AQI Value` : AQI value of Ozone of the city
- `Ozone AQI Category` : AQI category of Ozone of the city
- `NO2 AQI Value` : AQI value of Nitrogen Dioxide of the city
- `NO2 AQI Category` : AQI category of Nitrogen Dioxide of the city
- `PM2.5 AQI Value` : AQI value of Particulate Matter with a diameter of 2.5 micrometers or less of the city
- `PM2.5 AQI Category` : AQI category of Particulate Matter with a diameter of 2.5 micrometers or less of the city

```python
# Making the multicolored plot, each trace is a different pollutant
def make_country_stats(df, country_name):

    main_df = df[df["country_name"] == country_name]

    fig = go.Figure()
    fig.add_trace(go.Histogram(
        x=main_df["ozone_aqi_category"],
        name='Ozone AQI', # name used in legend and hover labels
        marker_color='#ef476f',
        opacity=0.75
    ))

    fig.add_trace(go.Histogram(
        x=main_df["co_aqi_category"],
        name='CO2 AQI', # name used in legend and hover labels
        marker_color='#ffd166',
        opacity=0.75
    ))

    fig.add_trace(go.Histogram(
        x=main_df["no2_aqi_category"],
        name='NO2 AQI', # name used in legend and hover labels
        marker_color='#06d6a0',
        opacity=0.75
    ))

    fig.add_trace(go.Histogram(
        x=main_df["pm2.5_aqi_category"],
        name='PM2.5 AQI', # name used in legend and hover labels
        marker_color='#118ab2',
        opacity=0.75
    ))

    fig.update_layout(
        title_text=str(country_name) + ' AQI Results', # title of plot
        xaxis_title_text='Pollutant Levels', # xaxis label
        yaxis_title_text='Count', # yaxis label
        bargap=0.2, # gap between bars of adjacent location coordinates
        bargroupgap=0.1, # gap between bars of the same location coordinates
    )

    return fig

# Making the miniplots
def subplot(df, country_name, pollutant_type, category, color):

    main_df = df[df["country_name"] == country_name]

    fig = go.Figure()
    fig.add_trace(go.Histogram(
        x=main_df[category],
        name=pollutant_type, # name used in legend and hover labels
        marker_color=color,
        opacity=0.75
    ))
    fig.update_layout(
        title_text=pollutant_type,
        autosize=True
    )

    return fig
```

```python
def total_cate(row):
    return help.loc[row]["co_aqi_category"] + help.loc[row]["ozone_aqi_category"] + help.loc[row]["no2_aqi_category"] + help.loc[row]["pm2.5_aqi_category"]

help["Total Category"] = help.index.map(total_cate)
help = help.sort_values(by="Total Category")
help

top_city_indexes = help.groupby('country_name')['Total Category'].idxmax()
bottom_city_indexes = help.groupby('country_name')['Total Category'].idxmin()

betty = help.loc[top_city_indexes]
crocker = help.loc[bottom_city_indexes]

betty = betty.sort_values(by='country_name', ascending=True)
crocker = crocker.sort_values(by='country_name', ascending=True)
```

```python
# Plotting Methods for the bar method overview of the city
def get_city_name(city):
    city_name = city.split(" - ")[-1]
    return city_name

def choose_plot(df, city_name, color, worst_best, plot_type):

    if plot_type == "Bar Plot":
        return make_city_bar_stats(df, city_name, color, worst_best)
    elif plot_type == "Radar Plot":
        return make_city_radar_stats(df, city_name, color, worst_best)

def make_city_bar_stats(df, city_name, color, worst_best):

    city_name = get_city_name(city_name)

    if worst_best == "Best":
        df = betty
    elif worst_best == "Worst":
        df = crocker

    df = df.set_index("city_name")
    df = df.drop(columns=["country_name","aqi_category","aqi_value","co_aqi_category","ozone_aqi_category","no2_aqi_category","pm2.5_aqi_category", "Total Category"])
    df = df.loc[[city_name]]
    df["City"] = df.index
    df = df.melt(id_vars=["City"], var_name=['Pollutant'], value_name="Total")

    fig = go.Figure()
    fig.add_trace(go.Bar(
        x=df['Pollutant'],
        y=df['Total'],
        name=city_name,
        marker_color=color
    ))

    return fig

def make_city_radar_stats(df, city_name, color, worst_best):

    if worst_best == "Best":
        wayne = betty
    elif worst_best == "Worst":
        wayne = crocker

    city_name = get_city_name(city_name)
    wayne = wayne.set_index("city_name")

    df = pd.DataFrame(dict(
        r=[wayne.loc[city_name]["co_aqi_category"],
            wayne.loc[city_name]["no2_aqi_category"],
            wayne.loc[city_name]["ozone_aqi_category"],
            wayne.loc[city_name]["pm2.5_aqi_category"]],
        theta=['CO2 category','NO2 Category',
              'Ozone Category', 'PM 2.5 Category']))

    fig = px.line_polar(df, r='r', theta='theta', line_close=True)

    fig.update_layout(polar=dict(radialaxis=dict(range=[0,5])))

    fig.update_traces(fill='toself', fillcolor=color, line_color=color, opacity=.75)

    return fig
```

```python
# Getting the list of the best/worst cities countries in terms of aqi

best_cities = []
for index, row in betty.iterrows():
    city = row["city_name"]
    country = row["country_name"]
    best_cities.append(str(country) + " - " + str(city))

worst_cities = []
for index, row in crocker.iterrows():
    city = row["city_name"]
    country = row["country_name"]
    worst_cities.append(str(country) + " - " + str(city))
```

```python
################################################################
# Making the Drop down selections

best_city1 = pnw.Select(options=best_cities, name="Countries' Best Cities");
best_city2 = pnw.Select(options=best_cities, name="Countries' Best Cities");
worst_city1 = pnw.Select(options=worst_cities, name="Countries' Worst Cities");
worst_city2 = pnw.Select(options=worst_cities, name="Countries' Worst Cities");
bar_radar = pnw.RadioButtonGroup(options=["Bar Plot", "Radar Plot"], name="Bar or Radar Plot");

################################################################

best1 = pn.bind(
    choose_plot,
    df = betty,
    city_name = best_city1,
    color = '#e07a5f',
    worst_best = "Best",
    plot_type = bar_radar
)

best2 = pn.bind(
    choose_plot,
    df = betty,
    city_name = best_city2,
    color = '#81b29a',
    worst_best = "Best",
    plot_type = bar_radar
)

worst1 = pn.bind(
    choose_plot,
    df = betty,
    city_name = worst_city1,
    color = '#5C91BA',
    worst_best = "Worst",
    plot_type = bar_radar
)

worst2 = pn.bind(
    choose_plot,
    df = betty,
    city_name = worst_city2,
    color = '#3d405b',
    worst_best = "Worst",
    plot_type = bar_radar
)

column1 = pn.Column(best_city1,
                    best1,
                    worst_city1,
                    worst1)

column2 = pn.Column(best_city2,
                    best2,
                    worst_city2,
                    worst2
                    )

row1 = pn.Row(column1,
              column2,
              sizing_mode='stretch_width')

combined_columns = pn.Column("# City to City AQI Comparator",
                             "## How do the top and worst cities in each country compare to one?",
                             "The cities are chosen by having the highest or lowest total category score based by adding all of the category ratings together. You're given each country's best (... or worst) representative!")

city_v_city = pn.Column(combined_columns,
                                        bar_radar,
                                        row1
                                        )
city_v_city
```

```python
# Ploting the double y axis plot with a bar and line plot trace
def secondary_axes(df, by_country, city_limit, pollutant, pollutant2):
  fig = make_subplots(specs=[[{"secondary_y": True}]])

  fig.add_trace(
    go.Scatter(x=df[df['country_name'] == by_country]['city_name'].iloc[:city_limit], y=df[df['country_name'] == by_country][pollutant], name=pollutant, hovertext=df[df['country_name'] == by_country]),
    secondary_y=True,
  )
  fig.add_trace(
    go.Bar(x=df[df['country_name'] == by_country]['city_name'].iloc[:city_limit], y=df[df['country_name'] == by_country][pollutant2], name=pollutant2),
    secondary_y=False,
  )
  fig.update_layout(
    title_text="AQI Pollutant Contributions by " + pollutant
    , height = 750,
    width = 1250
  )
  fig.update_xaxes(title_text="City Names")
  fig.update_yaxes(title_text=pollutant, secondary_y=True)
  fig.update_yaxes(title_text="True AQI value", secondary_y=False)

  return fig
```

```python
data="/content/global-air-pollution-data/global_air_pollution_data.csv"
df_2 = pd.read_csv(data)
df = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/2014_world_gdp_with_codes.csv')
df = df.set_index("COUNTRY")
df

# List of countries
barbara = df_2.drop(columns=["city_name", "aqi_category","co_aqi_category","ozone_aqi_category","no2_aqi_category", "pm2.5_aqi_category"])
barbara = barbara.groupby(["country_name"]).mean()
countries = list(barbara.index)
barbara["country"] = countries

# Adding the country codes to our dataframe so Choropleth can plot the map properly
def get_code(country_name):
    if country_name in df.index:
        new_value = df.loc[country_name, "CODE"]
    else:
        new_value = 0
    return new_value

barbara["code"] = barbara["country"].apply(get_code)

# Fixing the data because some didn't match
barbara.loc["Bolivia (Plurinational State of)","code"] = "BOL"
barbara.loc["Congo","code"] = "COG"
barbara.loc["Czechia","code"] = "CZE"
barbara.loc["Côte d'Ivoire","code"] = "CIV"
barbara.loc["Democratic Republic of the Congo","code"] = "COD"
barbara.loc["Gambia","code"] = "GMB"
barbara.loc["Iran (Islamic Republic of)","code"] = "IRN"
barbara.loc["Kingdom of Eswatini","code"] = "SWZ"
barbara.loc["Lao People's Democratic Republic","code"] = "LAO"
barbara.loc["Myanmar","code"] = "MMR"
barbara.loc["Republic of Korea","code"] = "KOR"
barbara.loc["Republic of Moldova","code"] = "MDA"
barbara.loc["Republic of North Macedonia","code"] = "MKD"
barbara.loc["Russian Federation","code"] = "RUS"
barbara.loc["State of Palestine","code"] = "KNA"
barbara.loc["Syrian Arab Republic","code"] = "SYR"
barbara.loc["United Kingdom of Great Britain and Northern Ireland","code"] = "GBR"
barbara.loc["United Republic of Tanzania","code"] = "TZA"
barbara.loc["United States of America","code"] = "USA"
barbara.loc["Venezuela (Bolivarian Republic of)","code"] = "VEN"
barbara.loc["Viet Nam","code"] = "VNM"

barbara
```

```python
# Selecting only the pollutants columns
pollutants = ['aqi_value', 'co_aqi_value\t', 'no2_aqi_value', 'ozone_aqi_value', 'pm2.5_aqi_value']

barbara['text'] = barbara["country"] + '<br>' + \
    'CO2 AQI: ' + barbara['co_aqi_value\t'].astype(str) + '<br>' + 'NO2 AQI: ' + barbara['no2_aqi_value'].astype(str) + '<br>' + \
    'Ozone AQI: ' + barbara['ozone_aqi_value'].astype(str) + '<br>' + 'PM2.5 AQI: ' + barbara['pm2.5_aqi_value'].astype(str)

# Create a Select widget for pollutants
pollutant_select = pnw.Select(options=pollutants, name="Pollutant")
vmax = pnw.EditableFloatSlider(name='Value Maximum', start=0, end=500, step=1, value=200)
vmin = pnw.EditableFloatSlider(name='Value Minimum', start=0, end=500, step=1, value=0)
color_select = pnw.Select(options=["Viridis", "Plasma", "YlOrRd"], name="Color")

# Making the Choropleth plot method it uses the selection parameters to build the plot
def make_city_choro_stats(df, pollutant_title, vmax_slider, vmin_slider, color_palette):

    fig = px.choropleth(df,
                        locations="code",
                        color=df[pollutant_title],
                        hover_name=barbara['text'],
                        range_color=(vmin_slider, vmax_slider),
                        color_continuous_scale=color_palette
                    )

    fig.update_layout(
        title_text=f'Distribution of {pollutant_title} in Cities',
        geo=dict(showframe=False, showcoastlines=False),  # Hide frame and coastlines
        font=dict(color='black'),
        margin=dict(l=0, r=0, b=0, t=0),
        height=600,
        width=1200
    )

    return fig

# Assembling the Dashboard
choroplot = pn.bind(
    make_city_choro_stats,
    df=barbara,
    vmax_slider = vmax,
    vmin_slider = vmin,
    pollutant_title=pollutant_select,
    color_palette = color_select
)

combined_columns = pn.Column("# Pollutant AQI Map Viewer",
                             "## Which regions and countries show the highest pollutation values?",
                             "Select which pollutant you want to view. Customise the max and minimum color scale!")

combined_row = pn.Row(pollutant_select, color_select, vmax, vmin)

choropleth_plot = pn.Column(
    combined_columns,
    combined_row,
    choroplot
)

# Show the dashboard
choropleth_plot
```

```python
# Adding the continents
def find_cont(x):
  if x is not None and pd.isnull(x) == False:
    return continent_map[x]
  return None

data="/content/global-air-pollution-data/global_air_pollution_data.csv"
df = pd.read_csv(data)
df['continent'] = df.country_name.map(find_cont)
df
```

```python
# Doing groupby to average out the country stats to place into the continent
tmp_df = df.groupby(['country_name','continent'])[['aqi_value','co_aqi_value\t','ozone_aqi_value','no2_aqi_value','pm2.5_aqi_value']].mean()
tmp_df = tmp_df.reset_index()
fd = df.groupby('country_name')['city_name'].count()
tmp_df = tmp_df.merge(fd, left_on = 'country_name', right_index=True, how='left')
tmp_df.rename(columns={'city_name' : 'city_count'}, inplace=True)
tmp_df
```

```python
# Making the pie plot, with all the displaying details, catered to continents
def pie_vars(df, continent, pollutant):
  tmp2 = df[df['continent'] == continent]
  fig = {
    "data": [
        {
            "values":tmp2[pollutant],
            "labels": tmp2['country_name'],
            "domain": {"x": [0, .5]},
            "name": "Proportions of average AQI for cities in country",
            "text": tmp2['country_name'],
            "hovertemplate":"%{label}: <br> Percent %{percent} </br> # of cities: %{text}",
            "hole": .2,
            "type": "pie"
        },
    ],
    "layout": {
        "title": ("Proportions of average AQI for Countries in " + continent) if  continent is not None else " ",
        "annotations": [
        { "font": {"size": 10},
          "showarrow": False,
          "text": " ",
          "x" : 0.5,
          "y" : 0.8
        },
      ],
        "width": 1400,
        "height": 1100
    }
  }

  return fig
```

```python
continent_select = pnw.Select(options=list(df.continent.unique()), name="Continents")
pollutant_select = pnw.Select(options=['aqi_value','co_aqi_value\t','ozone_aqi_value','no2_aqi_value','pm2.5_aqi_value'], name="Pollutant")

plot = pn.bind(
    pie_vars,
    df=tmp_df,
    continent = continent_select,
    pollutant = pollutant_select
)

combined_columns = pn.Column("# Pie Chart AQI Display",
                             "## Which countries are contributing to a continent's aqi's values?",
                             "Select which countinent you want to view and then the pollutant!")

interactive_plot_type_options = pn.Row(continent_select,
                                       pollutant_select)
world = pn.Column(combined_columns,
                  interactive_plot_type_options,
                  plot)

world
```
